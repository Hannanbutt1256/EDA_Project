# Generalization of Price Column

Generalizing the `Price` column for better analysis. Converting arabs, crore lakhs and thousands into number format.

```python
def price_to_numeric(price_str):
    if pd.isnull(price_str):
        return None

    price_str = price_str.replace("PKR", "").strip()  # remove PKR and extra spaces

    # Handle units
    multiplier = 1
    if 'Arab' in price_str:
        multiplier = 1e9 # a shorthand way of writing  multiplier = 1000000000
        price_str = price_str.replace('Arab', '').strip()
    elif 'Crore' in price_str:
        multiplier = 1e7
        price_str = price_str.replace('Crore', '').strip()
    elif 'Lakh' in price_str:
        multiplier = 1e5
        price_str = price_str.replace('Lakh', '').strip()
    elif 'Thousand' in price_str:
        multiplier = 1e3
        price_str = price_str.replace('Thousand', '').strip()

    # Remove commas and convert to float
    try:
        return float(price_str.replace(',', '')) * multiplier
    except:
        return None

df['Price_numeric'] = df['Price'].apply(price_to_numeric)

```

# Generalization of Area Column


Generalizing the `Area` column to Marlas

```python
def area_to_marla(area_str):
    if pd.isnull(area_str):
        return None

    area_str = area_str.strip()

    # Initialize multiplier
    multiplier = 1

    if 'Sq. Yd' in area_str:
        multiplier = 1/30.25
        area_str = area_str.replace('Sq. Yd.', '').strip()
    elif 'Marla' in area_str:
        multiplier = 1
        area_str = area_str.replace('Marla', '').strip()
    elif 'Kanal' in area_str:
        multiplier = 20
        area_str = area_str.replace('Kanal', '').strip()

    # Convert number part to float and apply multiplier
    try:
        return round(float(area_str.replace(',', '')) * multiplier, 2)
    except:
        return None

df['Area_Marla'] = df['Area'].apply(area_to_marla)
```

As I was working with data i saw type of the property define whether it has floorsor not and the column of `Floors` has more than **40%** of values missing. So, instead of dropping it i do this instead.

If floors are null and their type is flat set floors to zero.

```python
# Fill Floors based on House Type logic
df.loc[(df['Floors'].isnull()) & (df['Type'] == 'Flat'), 'Floors'] = 0
df.loc[(df['Floors'].isnull()) & (df['Type'].isin(['Upper Portion', 'Lower Portion',
                                                    'Farm House', 'Room', 'Penthouse'])), 'Floors'] = 1
df.loc[(df['Floors'].isnull()) & (df['Type'].isin(['Room'])), 'Floors'] = 1

# For Houses specifically, use median floors of actual valid House listings
house_median = df.loc[df['Type'] == 'House', 'Floors'].median()
df.loc[(df['Floors'].isnull()) & (df['Type'] == 'House'), 'Floors'] = house_median
```

Used IQR to calculate the outliers and then printed them to check what method of outlier removal I should apply

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

for cols in numeric_cols:
    outliers = detect_outliers_iqr(df[cols])
    print(f"Outliers in {cols}:")
    print(outliers)
    print("\n")

    print("Largest values of an outlier")
    print(outliers.max())
    print("----------------------------------------------------------------------------------------------------------------")

```

```python
for city in top_cities:
    # Filter for current city
    city_df = df_treated[df_treated['City_1'] == city]

    # Get top 5 locations in this city (by number of listings or average price)
    top_locations = city_df['Location'].value_counts().nlargest(5).index

    # Filter dataframe for top locations
    city_top_df = city_df[city_df['Location'].isin(top_locations)]

    # Plot
    plt.figure(figsize=(8, 6))
    sns.barplot(
        data=city_top_df,
        x='Price_log',
        y='Location',
        hue='Type',
        palette='viridis'

    )
    plt.title(f'Price_log by House Type for Top 5 Locations in {city}')
    plt.xlabel('Location')
    plt.ylabel('Log of Price')
    plt.xticks(rotation=45)
    plt.legend(title='House Type')
    plt.show()
```

```python
# Filter for years 2021 to 2025
df_years = df_treated[df_treated['Built in year'].between(2021, 2025)]

# Mean price per year
year_price = df_years.groupby('Built in year')['Price_log'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=year_price, x='Built in year', y='Price_log', palette='viridis')
plt.title('Average Log of Price for Houses Built from 2021 to 2025')
plt.xlabel('Built-in Year')
plt.ylabel('Average Log of Price')
plt.show()
```

```python
plt.figure(figsize=(10,6))

# Filter out rows where Floors == 0
df_filtered = df_treated[df_treated['Floors'] > 0]

sns.barplot(x='Floors', y='Price_log', data=df_filtered)
plt.title('Price_log vs Number of Floors (Floors > 0)')
plt.xlabel('Number of Floors')
plt.ylabel('Log of Price')
plt.show()
```

```python
# Define luxury properties (top 25% by price)
price_threshold = df_treated['Price_numeric'].quantile(0.75)
df_treated['Luxury'] = df_treated['Price_numeric'] >= price_threshold

# Count of Servant Quarters in luxury vs non-luxury
luxury_counts = df_treated.groupby('Luxury')['Servant Quarters'].value_counts().unstack(fill_value=0)
print(luxury_counts)

# Visualization
luxury_counts.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')
plt.title('Servant Quarters in Luxury vs Non-Luxury Properties')
plt.xlabel('Luxury Property')
plt.ylabel('Count of Properties')
plt.show()
```
